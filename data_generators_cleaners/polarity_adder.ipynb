# The purpose of this notebook is to add polarity and sentiment to tweets

```python
#imports
import pandas as pd 
from datetime import datetime
from datetime import timedelta
# from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import os
```

## Load tweets

```python
#get btc tweets
cleaned = False
filename = "../datasets/general/BTCTWEETS.csv" if not cleaned else "../datasets/general/tweets_cleaned.csv"
tweet_data = pd.read_csv(filename, sep=';') if not cleaned else pd.read_csv(filename)
```

```python
tweet_data.head()
```

## Get sentiment

```python
#this is either textblob or vader
analyser = "vader"
```

```python
if(analyser == "TextBlob"):
    tweet_data['polarity'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else TextBlob(x).sentiment.polarity)
    tweet_data['subjectivity'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else TextBlob(x).sentiment.subjectivity)
elif(analyser == "vader"):
    sentiment_analyser = SentimentIntensityAnalyzer()
    tweet_data['compound'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else sentiment_analyser.polarity_scores(x)["compound"])
    tweet_data['pos_pol'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else sentiment_analyser.polarity_scores(x)["pos"])    
    tweet_data['neg_pol'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else sentiment_analyser.polarity_scores(x)["neg"])    
    tweet_data['neu_pol'] = tweet_data['text'].apply(lambda x: 'NaN' if pd.isnull(x) else sentiment_analyser.polarity_scores(x)["neu"])    
```

```python
tweet_data.head()
```

## Save to file

```python
folder = '../datasets/general/with_sentiment/vader';

if cleaned:
    folder = folder+"/cleaned"
    
if not os.path.exists(folder):
    os.makedirs(folder)
```

```python
tweet_data.to_csv(folder+"/tweets.csv")
```
